```python
import rdkit
import pandas as pd
import selfies as sf
from rdkit import Chem
from rdkit.Chem import PandasTools as pt
```

```python
sdf = pt.LoadSDF('ligprep_oral-bioavailability-out.sdf')
print(sdf.columns)
```

```python
sdf['smiles'] = sdf['ROMol'].apply(lambda x: Chem.MolToSmiles(x))
dataset = list(sdf['smiles'].values)
```

```python
selfies_dataset = list(map(sf.encoder, dataset))
max_len = max(sf.len_selfies(s) for s in selfies_dataset)
alphabet = sf.get_alphabet_from_selfies(selfies_dataset)
alphabet.add("[nop]")
alphabet = list(sorted(alphabet))
vocab_stoi = {symbol: idx for idx, symbol in enumerate(alphabet)}
vocab_itos = {idx: symbol for symbol, idx in vocab_stoi.items()}
```

```python
selfies_dataset_toflatonehot =  sf.batch_selfies_to_flat_hot(selfies_dataset, vocab_stoi, max_len)
```
