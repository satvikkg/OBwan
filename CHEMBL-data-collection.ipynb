```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
from IPython.display import SVG
```

```python
available_resources = [resource for resource in dir(new_client) if not resource.startswith('_')]
print(available_resources)
```

```python
document = new_client.document
target_query = document.search('oral bioavailability')
targets = pd.DataFrame.from_dict(target_query)
targets.head(5)
print(targets.shape)
```

```python
actdf = pd.DataFrame()
for doc in targets['document_chembl_id']:
    activity = new_client.activity
    act = activity.filter(document_chembl_id=doc).filter(standard_type='F')
    actdf_temp = pd.DataFrame.from_dict(act)
    actdf = actdf.append(actdf_temp, ignore_index=True)
actdf
```

```python
actdf.to_csv('oral bioavailibility.csv',index=False)
```
