```python
import selfies as sf
import pandas as pd
```

```python
df = pd.read_csv('preclinical-bioavalilability-chembl.csv')
df
```

```python
dataset = list(df['canonical_smiles'].values)
dataset
```

```python
selfies_dataset = list(map(sf.encoder, dataset))
selfies_dataset
```

```python
max_len = max(sf.len_selfies(s) for s in selfies_dataset)
max_len
```

```python
alphabet = sf.get_alphabet_from_selfies(selfies_dataset)
alphabet.add("[nop]")
alphabet = list(sorted(alphabet))
alphabet
```

```python
vocab_stoi = {symbol: idx for idx, symbol in enumerate(alphabet)}
vocab_itos = {idx: symbol for symbol, idx in vocab_stoi.items()}
```

```python
selfies_dataset_toflatonehot =  sf.batch_selfies_to_flat_hot(selfies_dataset, vocab_stoi, max_len)
print(len(selfies_dataset_toflatonehot))
print(type(selfies_dataset_toflatonehot))
```

```python
onehot_toselfies_dataset = sf.batch_flat_hot_to_selfies(selfies_dataset_toflatonehot, vocab_itos)
print(len(onehot_toselfies_dataset))
print(type(onehot_toselfies_dataset))
```

```python
decoded_dataset = []
for selfie in onehot_toselfies_dataset:
    decoded_dataset.append(sf.decoder(selfie))
len(decoded_dataset)
decoded_dataset
```
